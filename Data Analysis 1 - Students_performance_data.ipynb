```python
#comparison of gender and race/ethnicity
#crosstabulation

aa=pd.crosstab(da1['gender'], da1['race/ethnicity'])
aa
```

```python
#multiple bar diagram

aa.plot(kind='bar', cmap='copper')
```

```python
#stacked bar diagram

x = pd.crosstab(da1['gender'], da1['race/ethnicity'])
x.div(x.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = True, figsize = (4, 4))
```
